# bootstrapping data into larger dataset and splitting data into training and test data

```python
#Using resample function from sklearn to increase the dataset to 1000 rows
cars_new = resample(cars, n_samples=1000)
```

```python
#defining X and y variables for splitting data can be split into training set and test set
X1 = cars_new[['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway',
       'Weight', 'Wheelbase', 'Length']] #Creating dataset X1 with columns of interest from expanded cars dataset
y1 = cars_new['MSRP'] #Creating column 'MSRP' as output value
```

```python
#data is split into training and test set
X1_train, X1_test, y1_train, y1_test = train_test_split(X1_sc, y1_sc, test_size=0.20)
```

```python
#checking to see the dimensions of each of the newly formed datasets
X1_train.shape, X1_test.shape, y1_train.shape, y1_test.shape
```

```python
#Implementing Linear Regression function
lr = LinearRegression()
```

```python
#checking to see the cross validation score of dataset X1 with y1(output value - MSRP) using linear regression
scores_2 = cross_val_score(lr, X1, y1)
scores_2.mean()
```

```python
#fitting training data into Linear Regression Predictive algorithm
lr.fit(X1_train, y1_train)
```

```python
'''Measures prediction accuracy using test data. Apparently this predictive algorithm using Linear Regression will predict the 
correct value 70% of the time'''
lr.score(X1_test, y1_test)
```

```python
#predicted y values from test input values are stored in y1_pred
y1_pred = lr.predict(X1_test)
```

```python
#mean squared error is computed between predicted y values and actual y values. Turns out it is abnormally high
mean_squared_error(y1_test, y1_pred)
```

```python
#mean_absolute_error is computed between predicted y values and actual y values
mean_absolute_error(y1_test, y1_pred)
```
